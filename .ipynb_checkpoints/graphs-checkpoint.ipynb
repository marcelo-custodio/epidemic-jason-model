```python
import pandas as pd
```

```python
sim = pd.read_csv("run.csv")
sim[sim.columns] = sim[sim.columns].astype(str)
sim.head()
```

```python
N = len(sim.columns)
N
```

```python
count_S = 0
count_I = 0
count_R = 0
S = []
I = []
R = []
for row in sim.iloc:
    for ag in row:
        ag = ag.strip()
        if ag == "S":
            count_S += 1
        elif ag == "I":
            count_I += 1
        elif ag == "R":
            count_R += 1
    S.append(count_S)
    I.append(count_I)
    R.append(count_R)
    if (count_S + count_I + count_R) < N:
        print("Bug")
    count_S = 0
    count_I = 0
    count_R = 0
```

```python
sim_c = pd.DataFrame({"S":S/len(sim.col, "I":I, "R":R})
sim_c.head()
```

```python
sim_c.plot.line()
```
